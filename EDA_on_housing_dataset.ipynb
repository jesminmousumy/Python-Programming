```python
#import libraries
import pandas as pd  #data manipulation
import numpy as np  #data manipulation
import matplotlib.pyplot as plt  #data visualization
import seaborn as sns  #data visualization


#load the dataset
df = pd.read_csv('advanced_housing_dataset.csv')

#display the first few rows of the dataframme
print(df.head())

#clean data
df.dropna(inplace=True)

#scatter plot: GrLivArea  vs SalePrice
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='GrLivArea', y='SalePrice')
plt.title('Relationship between Living Area and Sale Price')
plt.xlabel('Above Living Area(sqft)')
plt.ylabel('SalePrice($)')
plt.show()

#Histogram Sale Prices
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='SalePrice', bins=20, kde=True)
plt.title('Distribution of Sale Prices')
plt.xlabel('SalePrice ($)')
plt.ylabel('Frequency')
plt.show()

#Box plot : Sale Prices by overall quality
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='OverallQual', y='SalePrice')
plt.title('Sale Prices by Overall Quality')
plt.xlabel('Overall Quality')
plt.ylabel('SalePrice ($)')
plt.show()

#Pairplot of numerical features
sns.pairplot(df[['SalePrice', 'GrLivArea', 'OverallQual', 'YearBuilt']])
plt.show()

#Heatmap of Correlation Matrix
plt.figure(figsize=(10,6))
correlation_matrix = df.corr()
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()




```
