```python
#import necessary libraries
import pandas as pd   #used for data manipulation and analysis
import matplotlib.pyplot as plt #used for creating static, interactive and animated visualizations
import seaborn as sns      #high-level interface for drawing statistical graphics

df= pd.read_csv('uber_pickups.csv')   #reads csv file and loads it into a pandas Dataframe
print(df.head())   #display the first 5 rows of the dataset


#clean the data by dropping missing values
df.dropna(inplace=True)   #removes rows that contain missing values and the changes are made directly in original

#convert pick up datetime to a usable format
df['pickup_time'] = pd.to_datetime(df['pickup_time'])

#extract the hour from the pickup time for analysis
df['pickup_hour'] = df['pickup_time'].dt.hour  #extract the hours from each timestamp in the pickup

#create a  histogram to visualize the distribution of pickups by hour
sns.histplot(df['pickup_hour'], bins=24, kde=False) #create a histogram using the columns 'hour. from DataFrame df
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Pickups')
plt.title('Distribution of uber Pickups by Hour')
plt.show()

#create a scatter plot to visualize the relationship between pickup time and number of rides
plt.scatter(df['pickup_hour'], df['pickup_time']) #create a scatter plot where each plot point represents a specific hour an
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Pickups')
plt.title('Relationship between Pickup Time and Number of Rides')
plt.show()



```
